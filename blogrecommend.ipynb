```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
import re


names = []
rtypes = []
times = []
addrs = []
inss = []
url = "https://www.tutorcircle.hk/blog/30240/%E8%B7%B3%E8%88%9Estudio-%E5%88%9D%E5%AD%B8%E8%80%85-%E8%B7%B3%E8%88%9Estudio%E6%8E%A8%E8%96%A6/#1_%E8%B7%B3%E8%88%9Estudio%E7%B0%A1%E4%BB%8B"
quote_page = requests.get(url) 
# print("\n***** Page " + str(i) +" in action *****")
soup = BeautifulSoup(quote_page.content,'html.parser')
for item in soup.findAll("h3"):
    name=item.find("strong")
    if name != None:
        names.append(name.text)
    else:
        continue
    rtypeInd = 0
    for ind, info in enumerate(item.next_siblings):#h3, 
        if hasattr(info, "text") and re.match(u"舞蹈類型.*", info.text):
            rtypeInd = ind
            break
    for ind, info in enumerate(item.next_siblings):
        if ind == rtypeInd:
            rtype = info
            if rtype != None:
                rtypes.append(rtype.text)
            else:
                rtypes.append(None)
        if ind == rtypeInd+1:
            time = info
            if time != None:
                times.append(time.text)
            else:
                times.append(None)
        if ind == rtypeInd+2:
            addr = info
            if addr != None:
                addrs.append(addr.text)
            else:
                addrs.append(None)
        if ind == rtypeInd+3:
            ins = info.find("a")
            if ins and ins['href']:
                inss.append(ins['href'])
                break
        if ind == rtypeInd+4:
            ins = info.find("a")
            if ins != None:
                inss.append(ins['href'])
            else:
                inss.append(None)

                
df=pd.DataFrame({"Dance Name":names, "Room Type":rtypes, "Time":times, "ADDR":addrs, "INS":inss})           
df.to_csv('blogrecommend.csv', sep='\t', encoding='utf-8')
df
    
```
