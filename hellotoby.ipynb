```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd

url="https://www.hellotoby.com/zh-hk/s/%E8%B7%B3%E8%88%9E/?isMobile=true"
response = requests.get(url)
htmlcontent = response.content
soup = BeautifulSoup(htmlcontent,"html.parser")
print(soup.prettify)

```

```python
studio=[]
locations=[]
ratings=[]
#studio=soup.find('a',attrs={'class':'chxpzi-5 dbKwvP'})
#print(studio.text)
for i in range(0,17): # Page counter
    url = "https://www.hellotoby.com/zh-hk/s/%E8%B7%B3%E8%88%9E/?isMobile=true&offset="+str(i*10)+"&storeType=O_DAC"
    quote_page = requests.get(url) 
    # print("\n***** Page " + str(i) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    for a in soup.findAll('div',attrs={'class':'chxpzi-0 jvicRm'}):
         page_no = str(i)      
         name=a.find('a',attrs={'class':'chxpzi-5 dbKwvP'})
         location=a.find('div',attrs={'class':'chxpzi-9 bmKUmC'})
         rating=a.find('span',attrs={'class':'gycz1z-1 eYxvBj'})       
         if name != None:
            if name != None:
                studio.append(name.text)
            else:
                ratings.append(None) 
            if location != None:
                locations.append(location.text)
            else:
                ratings.append(None) 
            if rating != None: 
                ratings.append(rating.text)
            else:
                ratings.append(None) 
                
       

print(studio)
print(locations)
print(ratings)
```

```python
df=pd.DataFrame({"studio":studio,"location":locations, "Ratings":ratings})           
df 
```

```python
df=pd.DataFrame({"studio":studio,"location":locations,"Ratings":ratings})  
df.to_csv('hellotoby.csv', sep='\t', encoding='utf-8')
df
```
