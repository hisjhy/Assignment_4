```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd

url="https://classpass.com/search/hong-kong/dance-classes/5AzQs7EsJYw"
response = requests.get(url)
htmlcontent = response.content
soup = BeautifulSoup(htmlcontent,"html.parser")
print(soup.prettify)
```

```python
studio=[]
locations=[]
ratings=[]

for a in soup.findAll('li',attrs={'class':'_3vk1F9nlSJQIGcIG420bsK'}):
         #page_no = str(i)      
         name=a.find('a',attrs={'class':'_1a94aFGSes-pROpp23ORnK'})
         location=a.find('div',attrs={'class':'_3oAhZ5phwNNDu5r1BFcnt1 text--ellipsis'})
         rating=a.find('span',attrs={'class':'ratings__rating ratings--child'})
        
         if name != None:
            if name != None:
                studio.append(name.text)
            else:
                ratings.append(None) 
            if location != None:
                locations.append(location.text)
            else:
                ratings.append(None) 
            if rating != None: 
                ratings.append(rating.text)
            else:
                ratings.append(None) 
           
      

print(studio)
print(locations)
print(ratings)



```

```python
print(len(studio))
print(len(locations))
print(len(ratings))

```

```python
df=pd.DataFrame({"studio":studio,"location":locations,'ratings':ratings})  
df.to_csv('clssspass.csv', sep='\t', encoding='utf-8')
df
```
